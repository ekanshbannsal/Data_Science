```
# pallindrome
s = input("enter a string ")
l = len(s)
is_pallindrome = True
for i in range (l//2) :
  if s[i] != s[l-1-i] :
    is_pallindrome = False
    break

if is_pallindrome :
  print("it is pallindrome")
else :
  print("it is not a pallindrome")
```
